Cleaning the Gender coloumn

```python
patients["gender"].unique()
```

```python
# Standardization being carried out
patients["gender"] = patients["gender"].replace({
    "M":"Male",
    "F":"Female",
    "male":"Male",
    "female":"Female"
})
```

```python
# Filling missing gender nan
patients["gender"] = patients["gender"].fillna("Unknown")
```
